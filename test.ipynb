```python
import re 
f = open("intent.txt", "r")
intent_dict = {}
last_intent = ''

for x in f:
    if '## intent:' in x:
        intent = x.split(':')[-1].strip()
        if intent not in intent_dict:
            intent_dict[intent] = set()
        last_intent = intent
    else:
        utterance = x.strip('-').strip()
        utterance = re.sub(r'\([^)]*\)', '', utterance).replace('[', '').replace(']', '')
        utterance = re.sub(r'{.*?}', '', utterance).replace('{', '').replace('}', '')
        if utterance!= '':
            if len(utterance.split()) > 2:
                intent_dict[intent].add(utterance)
intent_dict
```

```python
from itertools import permutations
import random

my_training_list = []

for key in intent_dict:
    my_utterance_list = list(intent_dict[key])
    for entry in permutations(my_utterance_list, 2):
        my_training_list.append(entry)
# my_training_list[:10]
random.shuffle(my_training_list)
my_training_list[:10]
```

```python
from textattack.augmentation import EmbeddingAugmenter, WordNetAugmenter
embedding_aug = EmbeddingAugmenter(transformations_per_example=10, pct_words_to_swap=0.25)
wordnet_aug = WordNetAugmenter(transformations_per_example=10, pct_words_to_swap=0.25)

with open('gen_textattack.txt', 'w') as f:
    for key in list(intent_dict.keys()):
        my_utterance_list = list(intent_dict[key])
        for entry in my_utterance_list:
            f.write('## intent: '+entry + '\n')
            emb_augmentations = embedding_aug.augment(entry)
            wn_augmentations = wordnet_aug.augment(entry)
            total_augmentations = set(emb_augmentations+wn_augmentations)
            for item in set(total_augmentations):
                f.write(item + '\n')

comparison =[
    'A man with luggage on wheels standing next to a white van.',
    'A table full of vegetables and fruits piled on top of each other.',
    'Which is the best training institute in Pune for digital marketing and why?',
    'What are the best ways to lose weight?'
]         
for item in comparison:
    print(embedding_aug.augment(item))
    print(wordnet_aug.augment(item))
```

```python
with open('original.txt', 'w') as a, open('paraphrase.txt', 'w') as b:
    for sentence_pair in my_training_list:
        a.write(sentence_pair[0]+'\n')
        b.write(sentence_pair[1]+'\n')
```

```python
words = set()
mean_length = 0
    
for sentence_pair in my_training_list:
    mean_length += len(sentence_pair[0].split())
    mean_length += len(sentence_pair[1].split())
    words.update(sentence_pair[0].split())
    words.update(sentence_pair[1].split())

mean_length = mean_length/(len(my_training_list)*2)
print(len(words), mean_length)
```
